```python
import plotly.express as px

# حساب عدد مرات ظهور كل شعور
emotion_counts = df['emotion'].value_counts()

# تحديد أكثر 20 شعورًا فقط
top_emotions = emotion_counts[:20]

# إنشاء المخطط التفاعلي بحجم أكبر
fig = px.bar(
    x=top_emotions.index, 
    y=top_emotions.values, 
    title="Top 20 Emotions in the Dataset",
    labels={"x": "Emotion", "y": "Count"},
    color=top_emotions.values,  # تلوين حسب القيم
    color_continuous_scale="viridis",  # استخدام نفس التدرج اللوني
)

# تحسين شكل المحاور والتخطيط العام
fig.update_layout(
    xaxis_title="Emotion",
    yaxis_title="Count",
    xaxis_tickangle=-45,  # إمالة النصوص
    template="plotly_white",
    width=1200,  # توسيع العرض
    height=700   # توسيع الارتفاع
)

# عرض المخطط التفاعلي
fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# إضافة عمود لطول كل حوار
df['dialogue_length'] = df['empathetic_dialogues'].apply(lambda x: len(str(x)))

# اختيار فقط أكثر 15 شعور تكرارًا
top_15_emotions = df['emotion'].value_counts().index[:15]
df_filtered = df[df['emotion'].isin(top_15_emotions)]

# رسم Box Plot لطول الحوارات حسب المشاعر
plt.figure(figsize=(14, 7))
sns.boxplot(x='emotion', y='dialogue_length', data=df_filtered, palette="coolwarm")

# تحسين تنسيق الرسم البياني
plt.xlabel("Emotion", fontsize=14)
plt.ylabel("Dialogue Length", fontsize=14)
plt.title("Dialogue Length Distribution per Emotion", fontsize=16)
plt.xticks(rotation=30, ha="right", fontsize=12)  # إمالة النصوص لتكون واضحة
plt.grid(axis="y", linestyle="--", alpha=0.7)  # تحسين الشبكة

# عرض المخطط
plt.show()
```

```python
import plotly.express as px

# حساب نسبة كل شعور
emotion_counts = df['emotion'].value_counts().reset_index()
emotion_counts.columns = ['emotion', 'count']

# رسم الـ Treemap
fig = px.treemap(
    emotion_counts, 
    path=['emotion'], 
    values='count', 
    title="Emotion Distribution (Treemap)",
    color='count',
    color_continuous_scale='viridis'
)

fig.show()
```
